# Test technique

L'objectif de ce test consiste à prédire, à l'aide de modèles de régressions, le prix d'une habitation en fonction de variables pertinentes représentant différentes caractéristiques architecturales, géographiques et propres au voisinage.

Le jeu de données utilisé contient de nombreux attributs permettant de décrire 1460 habitations en fonction de plusieurs caractéristiques ainsi qu'une variable cible: 'SalePrice' contenant le prix de vente du bien en question.

Le jeu de données est à lire dans le fichier <em>"house_price.csv"</em> (les colonnes sont séparées par des ';')

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
## à compléter
from pandas.api.types import is_numeric_dtype
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate

# modeles regression
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import seaborn as sns
```

```python
## Insérez votre code ici
df = pd.read_csv('house_price.csv')
print(df.shape)
# df.head()
# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns',None)
display(df)
```

1460 echantillons et 81 features 

## Prétraitement

Faire les actions nécessaires pour rendre le jeu de données propre :
- traiter les valeurs manquantes / les doublons ;
- centrer-réduire ; 
- transformer les variable catégorielles en variables indicatrices ;
- séparer la variable cible du reste des variables ;
- séparer les données en un ensemble d'apprentissage et un ensemble de test ;

### VALEURS MANQUANTES / DOUBLONS

```python
## Insérez votre code ici

# df.duplicated().value_counts() # pas de doublons 
# df.info()

na80 = np.where(df.isna().sum()/df.shape[0] * 100 > 80)  # je prend l'index des colonnes qui ont plus de 80% de Nan 
print(len(na80[0]))
df = df.drop(df.columns[na80[0]] , axis = 1)
print(df.shape)

na25 = np.where(df.isna().sum()/df.shape[0] * 100 < 25)  # je prend l'index des colonnes qui ont moins de 25% de Nan 
for i in na25[0]:
    if (df.iloc[:,i].dtype in [np.float64, np.int64] ):
        mean = df.iloc[:,i].mean()
        df.iloc[:,i] = df.iloc[:,i].fillna(mean) # je remplace par la moyenne 
    
    if (df.iloc[:,i].dtype == object):
        index_max = df.iloc[:,i].value_counts().idxmax() # je remplace par l'object principal
        df.iloc[:,i] = df.iloc[:,i].fillna(index_max)
```

```python
# display(df)
```

### CENTRER REDUIRE

```python
## Insérez votre code ici

# Sélectionner les colonnes numériques
numeric_columns = df.select_dtypes(['int', 'float']).columns
# print(numeric_columns.shape)

# Centrer les valeurs de ces colonnes en soustrayant la moyenne de chaque colonne
df[numeric_columns] = df[numeric_columns].sub(df[numeric_columns].mean())

# Réduire les valeurs de ces colonnes en les divisant par l'écart-type de chaque colonne
df[numeric_columns] = df[numeric_columns].div(df[numeric_columns].std())
```

```python
# df.head()
```

### VARIABLES CATEGORIELLES EN VARIABLES INDICATRICES

```python
## Insérez votre code ici
object_columns = df.select_dtypes(['object']).columns

df = pd.get_dummies(df, columns=object_columns)

display(df)
```

### VARIABLE CIBLE

```python
## Insérez votre code ici
target = df['SalePrice'].to_numpy()
print(target)

features = df.drop(['SalePrice'], axis = 1).to_numpy()
print(features.shape)
```

### TRAIN / TEST

```python
## Insérez votre code ici
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.2 )
```

```python
print(X_train.shape)
print(y_train.shape)
```

## Apprentissage

Comparer quelques modèles différents par validation croisée sur le jeu de train.

```python
# Linear regression 
linear_regression = LinearRegression()
linear_regression.fit(X_train,y_train)

# scores = cross_validate(linear_regression,
#                        X_train,
#                        y_train,
#                        cv = 5)
# print(scores['test_score'])
acc = linear_regression.score(X_train,y_train)
print(acc)
```

```python
# LASSO
lasso_regression = Lasso(alpha=0.1)
lasso_regression.fit(X_train,y_train)

# scores = cross_validate(linear_regression,
#                        X_train,
#                        y_train,
#                        cv = 5)
# print(scores['test_score'])

acc_lasso = lasso_regression.score(X_train,y_train)
print(acc_lasso)
# print(lasso_regression.coef_)
```

```python
# RIDGE
ridge_regression = Ridge(alpha=0.1)
ridge_regression.fit(X_train,y_train)

# scores = cross_validate(linear_regression,
#                        X_train,
#                        y_train,
#                        cv = 5)
# print(scores['test_score'])

acc_ridge = ridge_regression.score(X_train,y_train)
print(acc_ridge)
# print(ridge_regression.coef_)
```

```python
# Elasticnet
elasticnet_regression = ElasticNet(alpha=0.1)
elasticnet_regression.fit(X_train,y_train)

# scores = cross_validate(linear_regression,
#                        X_train,
#                        y_train,
#                        cv = 5)
# print(scores['test_score'])

acc_elasticnet = elasticnet_regression.score(X_train,y_train)
print(acc_elasticnet)
# print(ridge_regression.coef_)
```

```python
# SVR
svr_regression = SVR()
svr_regression.fit(X_train,y_train.ravel())

# scores = cross_validate(linear_regression,
#                        X_train,
#                        y_train,
#                        cv = 5)
# print(scores['test_score'])

acc_elasticnet = svr_regression.score(X_train,y_train)
print(acc_elasticnet)
```

```python
# Arbre de decision
arbre_regression = DecisionTreeRegressor()
arbre_regression.fit(X_train,y_train)

# scores = cross_validate(linear_regression,
#                        X_train,
#                        y_train,
#                        cv = 5)
# print(scores['test_score'])

acc_arbre = arbre_regression.score(X_train,y_train)
print(acc_arbre)
```

Pour le meilleur modèle, afficher les variables les plus importantes du modèle, relativement aux coefficients calculés.

```python
coefficients = linear_regression.coef_

dico = {}

for coefficient, feature_name in zip(coefficients[0], df.columns):
#     print(f'{feature_name}: {coefficient}')
    dico[feature_name] = coefficient
# print(dico)
dico = dict(sorted(dico.items(), key=lambda item: item[1] , reverse = True))
print(dico)
```

Visualiser les corrélations entre les différentes variables du modèle.

```python
print(list(dico.keys())[:10])
```

```python
plt.figure(figsize = (25,25))
sns.heatmap(df[list(dico.keys())[:25]].corr() ,vmin = -0.6 ,annot = True)
```

Mesurer l'erreur de prédiction du modèle retenu sur le jeu de test.

```python
print(linear_regression.score(X_test,y_test))
```

```python
# Faire des prédictions sur le jeu de test
y_pred = linear_regression.predict(X_test)

# Mesurer l'erreur de prédiction avec l'erreur quadratique moyenne (MSE)
mse = mean_squared_error(y_test, y_pred)
print(mse)

# Mesurer l'erreur de prédiction avec l'erreur absolue moyenne (MAE)
mae = mean_absolute_error(y_test, y_pred)
print(mae)

r2 = r2_score(y_test, y_pred)
print(r2)
```
